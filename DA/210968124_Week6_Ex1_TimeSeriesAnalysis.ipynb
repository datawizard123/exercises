```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df=pd.read_csv('employment.csv')
df
```

### 1. Convert datestamp column to a datetime object and Set the datestamp columns as the index of your DataFrame. Check if there are missing values in each column.

```python
#converting to datetime object
df['datestamp']=pd.to_datetime(df['datestamp'])
df['datestamp']
```

```python
df.info()
```

```python
#setting datestamp as index
df=df.set_index(['datestamp'])
df.head()
```

```python
#checking for missing value
df.isnull().sum()
```

```python
#there are no missing values
```

```python
df.describe()
```

```python
#removing outliers
q1=df.quantile(0.25)
q2=df.quantile(0.50)
q3=df.quantile(0.75)
iqr=q3-q1
u=q3+1.5*iqr
l=q1-1.5*iqr
```

```python
df[df>u]=np.nan
df[df<l]=np.nan
df
```

```python
df.fillna(q2,inplace=True)
df
```

### 2. Generate a boxplot to find the distribution of unemployment rate for every industry.

```python
sns.boxplot(data=df,orient="h")
```

### 3. Using line chart Visualize the unemployment rate of workers by industry.

```python
df.plot(subplots=True,figsize=(10,40))
plt.show()
```

### 4. Plot the monthly and yearly trends.

```python
df_by_month = df.resample('M').mean()
df_by_year = df.resample('Y').mean()
start, end = '2001-01', '2010-02'
fig, ax = plt.subplots(figsize=(20, 15))
#monthly and yearly trend of agriculture
ax.plot(df_by_year.loc[start:end, 'Agriculture'], marker='.', linestyle='-', linewidth = 0.5, label='yearly', color='black')
ax.plot(df_by_month.loc[start:end, 'Agriculture'], marker='o', markersize=8, linestyle='-', label='monthly', color='coral')
ax.set_ylabel("Open")
ax.legend()
plt.show()
```

```python
#yearly trend of all industries
df_by_year.plot(figsize=(10,10), title = 'Yearly')
```

```python
#monthly trend of all industries
df_by_month.plot(figsize=(10,10), title = 'Monthly')
```

```python
#yearly and monthly trend of all industries
fig, axes = plt.subplots(15, 1, figsize = (10,33), sharex = True)
for name, ax in zip(['Agriculture', 'Business services', 'Construction', 'Durable goods manufacturing', 'Education and Health', 'Finance', 'Information', 'Leisure and hospitality', 'Manufacturing', 'Mining and Extraction', 'Nondurable goods manufacturing', 'Other', 'Self-employed', 'Transportation and Utilities', 'Wholesale and Retail Trade'], axes):
    ax.plot(df[name])
    ax.plot(df_by_year[name])
    ax.set_title(name)
```

### 5. Apply time series decomposition to your dataset to visualize the trend and seasonality.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
#trend
for i in df.columns:
    result=seasonal_decompose(df[i], model='multiplicable', period=12)
    result.trend.plot()
```

```python
#seasonalality(couldn't make separate graphs for each industry)
for i in df.columns:
    result=seasonal_decompose(df[i], model='multiplicable', period=12)
    result.seasonal.plot(figsize=(10,10))
```

### 6. Visualize the seasonality of Agriculture, Health and Finance sector.

```python
for i in ['Agriculture','Education and Health','Finance']:
    result=seasonal_decompose(df[i], model='multiplicable', period=12)
    result.seasonal.plot(figsize=(10,20))
plt.legend()
```

```python
df_month = df.resample("M").mean()
fig, ax = plt.subplots(figsize=(14,6))
ax.bar(df_month['2001':].index, df_month.loc['2001':, "Agriculture"], width=25, align='center',label = 'Agriculture')
ax.bar(df_month['2001':].index, df_month.loc['2001':, "Education and Health"], width=25, align='center',label = 'Education and Health')
ax.bar(df_month['2001':].index, df_month.loc['2001':, "Finance"], width=25, align='center',label = 'Finance')
plt.xlabel("Month")
plt.ylabel("Unemployment")
plt.legend()
```
