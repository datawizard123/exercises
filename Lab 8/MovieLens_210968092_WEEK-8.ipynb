# sklearn.metrics.pairwise_distances can be used to compute distance between all pairs of users. pairwise_distances() takes a metric parameter for what distance measure to use. Use cosine similarity for finding similarity among users.

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

# Find the 5 most similar user for user with user Id 25

```python
distance_matrix = pairwise_distances(newdf, metric='cosine')
```

```python
print('5 most similar users for user with user id 25 are: ')
print((distance_matrix[24]).argsort()[1:6]+1)
```
