### 4. For each of the numeric attributes (proteins up to vitamins), identify and replace all noisy data with the median of attribute.


```python
#quantiles for the numeric columns
q1=df.iloc[:,3:12].quantile(0.25) #25th percentile
q2=df.iloc[:,3:12].quantile(0.5) #median
q3=df.iloc[:,3:12].quantile(0.75) #75th percentile
```

```python
q1,q2,q3
```

```python
#inter quartile range
iqr=q3-q1
iqr
```

```python
#upper and lower bounds for outliers
u=q3+1.5*iqr
l=q1-1.5*iqr
u,l
```

```python
#identifying and replacing all outliers with nan
df[df>u]=np.nan
df[df<l]=np.nan
df
```

```python
#replacing all nan values with the median of the column
df.fillna(q2,inplace=True)
df
```

### 15. Design a Linear regression model to predict the rating of a cereal based on top 3 related nutrients. Tabulate the accuracy of the model using an 80, 20 split.

```python
#Top 3 related nutrients are sugars(negatively),protein,fat(negatively)
```

```python
#splitting the data into training and testing sets
from sklearn.model_selection import train_test_split
X=df[['protein','fat','sugars']]
y=df.iloc[:,15]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
#training the model on the training data
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
print(reg.score(X_test,y_test))
```

```python
#Accuracy of the model is 82.698% on the testing set
```

```python
#Scatter plots showing the predicted ratings v/s the actual ratings for each of
#the nutrients
```

```python
#predicted v/s actual ratings for protein
import matplotlib.pyplot as plt
y_pred = reg.predict(X_test)
plt.scatter(X_test.iloc[:,0], y_test, color ='b')
plt.scatter(X_test.iloc[:,0], y_pred, color ='k')
plt.show()
```

```python
#predicted v/s actual ratings for fat
y_pred = reg.predict(X_test)
plt.scatter(X_test.iloc[:,1], y_test, color ='b')
plt.scatter(X_test.iloc[:,1], y_pred, color ='k')
plt.show()
```

```python
#predicted v/s actual ratings for sugar
y_pred = reg.predict(X_test)
plt.scatter(X_test.iloc[:,2], y_test, color ='b')
plt.scatter(X_test.iloc[:,2], y_pred, color ='k')
plt.show()
```
