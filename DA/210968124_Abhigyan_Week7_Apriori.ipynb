```python
#importing necessary libraries
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#importing the dataset and converting it into a list
#using csv.reader instead of pd.read_csv because the dataset has empty values(i.e, rows of varying length), so we deal with it 
#by converting it into a list first
import csv
with open('groceries.csv',newline='') as groceries:
    csv_reader=csv.reader(groceries)
    data=list(csv_reader)
```

```python
#list converted to dataframe
df=pd.DataFrame(data)
df.head()
```

### 2. Prepare the data for finding association rules. Each transaction will contain a list of item in the transaction.

```python
data
```

### 3. Use Python library mlxtend and convert the transactions into a format that can be used in the Apriori method for finding frequent itemsets.

```python
#transforming the data where False indicates that the product is not present in the transaction and True indicates it is present
t=TransactionEncoder()
temp = t.fit(data).transform(data)
df = pd.DataFrame(temp, columns=t.columns_)
df
```

### 4. Find top selling items with minimum support of 2%.

```python
#top 10 selling single items of minimum support 0.02
temp1=apriori(df, min_support=0.02, use_colnames=True)
temp1['length']=temp1['itemsets'].apply(lambda x : len(x))
temp1[temp1['length']==1].sort_values(by=['support'],ascending=False).head(10)
```

### 5. Find all frequent itemsets with minimum support of 5%.

```python
#all frequent itemsets with minimum support 0.05
temp1[temp1['support']>=0.05].sort_values(by=['support'],ascending=False)
```

### 6. Find all frequent itemsets of length 2 with minimum support of 2%.

```python
#all frequent itemsets of length 2 with minimum support 0.02
temp1[(temp1['length']==2) & (temp1['support']>=0.02)].sort_values(by=['support'],ascending=False)
```

### 7. Find the top 10 association rules with minimum support of 2%, sorted by confidence in descending order.

```python
#top 10 association rules with minimum support 0.02, sorted by confidence in descending order
rules=association_rules(temp1,metric='confidence',min_threshold=0)
rules.sort_values(by=['confidence'],ascending=False).head(10)
```

### 8. Find association rules with minimum support of 2% and lift of more than 1.0.

```python
#association rules with minimum support 0.02 and minimum lift 1.0
rules1=association_rules(temp1,metric='lift',min_threshold=1.0)
rules1.sort_values(by=['lift'],ascending=False)
```
